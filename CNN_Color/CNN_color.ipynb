```python
from google.colab import drive
drive.mount('/content/drive/')

import numpy as np
import pandas as pd
import os, cv2
import re, random
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
import zipfile
from google.colab import files
from tensorflow.keras.callbacks import TensorBoard
from keras.callbacks import ModelCheckpoint
```

```python

fruits = ["Apple","Banana","Bellpepper","Carrot","Cucumber","Grape","Guava","Jujube","Mango","Orange","Pomegranate","Potato","Strawberry","Tomato"]

for fruit in fruits:
  path = f"/content/drive/MyDrive/IA_proyect/{fruit}.zip"
  extract_path = f"/content/drive/MyDrive/IA_proyect/{fruit}/ex_images"
  os.makedirs(extract_path, exist_ok=True)

  with zipfile.ZipFile(path, 'r') as zip_ref:
      zip_ref.extractall(extract_path)

  print(os.listdir(extract_path))

  category = os.listdir(extract_path)
  images = []
  labels = []

  columns = category
  Labelss = category

  count = 0
  for folder in columns:
      count2 = 0
      print('Loading --> ', folder)
      for img in os.listdir(os.path.join(extract_path,folder)):
          image = cv2.imread(os.path.join(extract_path,folder,img))
          image = cv2.resize(image, (128, 128))
          images.append(image)
          labels.append(count)
          count2 += 1
      print('    Total of --> ', count2)
      count += 1

  print(len(images))
  print(len(labels))


  unique_labels = list(set(labels))
  print(unique_labels)

  training_data = []
  for img, label in zip(images, labels):
      training_data.append([img, label])

  X = []
  y = []

  for img , lab in training_data:
      X.append(img)
      y.append(lab)

  X = np.array(X).astype(float)/255
  y = np.array(y)
  X.shape

  model_CNN = tf.keras.models.Sequential([
      tf.keras.layers.Conv2D(32, (3,3), activation ='relu', input_shape =(128,128,3)),
      tf.keras.layers.MaxPooling2D(2,2),
      tf.keras.layers.Conv2D(64, (3,3), activation ='relu'),
      tf.keras.layers.MaxPooling2D(2,2),
      tf.keras.layers.Conv2D(128, (3,3), activation ='relu'),
      tf.keras.layers.MaxPooling2D(2,2),

      tf.keras.layers.Dropout(0.5),
      tf.keras.layers.Flatten(),
      tf.keras.layers.Dense(250, activation='relu'),
      tf.keras.layers.Dense(2, activation='softmax'),
  ])

  model_CNN.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

  TensorBoardCNN = TensorBoard(log_dir='logs/denso')

  model_CNN.fit(X,y, batch_size= 32, validation_split=0.15, epochs=100, callbacks=[TensorBoardCNN])

  plt.figure(figsize=(10, 10))
  plt.xlabel("# Epochs")
  plt.ylabel("Loss")
  plt.plot(model_CNN.history.history['loss'])
  plt.savefig(f"loss_plot_{fruit}.png")

  files.download(f"/content/loss_plot_{fruit}.png")

  model_CNN.save(f"model_CNN_{fruit}.h5")

  files.download(f"/content/model_CNN_{fruit}.h5")
```

```python
#one model for all the fruits

fruits = ["Apple","Banana","Bellpepper","Carrot","Cucumber","Grape","Guava","Jujube","Mango","Orange","Pomegranate","Potato","Strawberry","Tomato"]
random_path = f"/content/drive/MyDrive/IA_proyect/Random/data"

# inicializate the images and labels
images = []
labels = []

# To charge the fruits in the envirment with label = "1"
for fruit in fruits:
  extract_path = f"/content/drive/MyDrive/IA_proyect/{fruit}/ex_images"

  print(os.listdir(extract_path))
  folders = os.listdir(extract_path)
  label_fruit = 1

  for folder in folders:
      count2 = 0
      print('Loading --> ', folder)
      for img in os.listdir(os.path.join(extract_path,folder))[0:167]:
          image = cv2.imread(os.path.join(extract_path,folder,img))
          image = cv2.resize(image, (128, 128))
          images.append(image)
          labels.append(label_fruit)
          count2 += 1
      print('    Total of --> ', count2)

print('Loading fruits is over')
# To charge the random in the envirment with label = "0"

print('Loading random start')
extract_path = random
label_fruit = 0
count2 = 0

random_path

for img in os.listdir(random_path):
    image = cv2.imread(os.path.join(random_path,img))
    image = cv2.resize(image, (128, 128))
    images.append(image)
    labels.append(label_fruit)
    count2 += 1
print('    Total of --> ', count2)


print(len(images))
print(len(labels))


unique_labels = list(set(labels))
print(unique_labels)

training_data = []
for img, label in zip(images, labels):
    training_data.append([img, label])

X = []
y = []

for img , lab in training_data:
    X.append(img)
    y.append(lab)

X = np.array(X).astype(float)/255
y = np.array(y)
print("The shape of X is: ", X.shape)

model_CNN = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(32, (3,3), activation ='relu', input_shape =(128,128,3)),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Conv2D(64, (3,3), activation ='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Conv2D(128, (3,3), activation ='relu'),
    tf.keras.layers.MaxPooling2D(2,2),

    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(250, activation='relu'),
    tf.keras.layers.Dense(2, activation='softmax'),
])

model_CNN.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

TensorBoardCNN = TensorBoard(log_dir='logs/denso')

model_CNN.fit(X,y, batch_size= 32, validation_split=0.15, epochs=100, callbacks=[TensorBoardCNN])

plt.figure(figsize=(10, 10))
plt.xlabel("# Epochs")
plt.ylabel("Loss")
plt.plot(model_CNN.history.history['loss'])
plt.savefig(f"loss_plot_fruit_random.png")

files.download(f"/content/loss_plot_fruit_random.png")

model_CNN.save(f"model_CNN_fruit_random.h5")

files.download(f"/content/model_CNN_fruit_random.h5")

# Convertir la historia en un DataFrame de pandas
history = model_CNN.history.history
df = pd.DataFrame(history)

# Guardar el DataFrame como un archivo CSV
df.to_csv('historia_entrenamiento.csv', index=False)


# Grfica de la prdida
plt.plot(history['loss'])
plt.plot(history['val_loss'])
plt.title('Prdida del Modelo')
plt.ylabel('Prdida')
plt.xlabel('poca')
plt.legend(['Entrenamiento', 'Validacin'], loc='upper left')
plt.savefig('perdida_modelo_fruit_random.png')
plt.show()

# Grfica de la precisin (si es aplicable)
if 'accuracy' in history:
    plt.plot(history['accuracy'])
    plt.plot(history['val_accuracy'])
    plt.title('Precisin del Modelo')
    plt.ylabel('Precisin')
    plt.xlabel('poca')
    plt.legend(['Entrenamiento', 'Validacin'], loc='upper left')
    plt.savefig('precision_modelo_fruit_random.png')
    plt.show()
```
