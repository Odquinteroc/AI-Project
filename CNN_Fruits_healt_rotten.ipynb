```python
import numpy as np
import os, cv2
import re, random
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
import zipfile
from google.colab import files
from tensorflow.keras.callbacks import TensorBoard

fruits = ["Apple","Banana","Bellpepper","Carrot","Cucumber","Grape","Guava","Jujube","Mango","Orange","Pomegranate","Potato","Strawberry","Tomato"]

for fruit in fruits:
  path = f"/content/drive/MyDrive/IA_proyect/{fruit}.zip"
  extract_path = f"/content/drive/MyDrive/IA_proyect/{fruit}/ex_images"
  os.makedirs(extract_path, exist_ok=True)

  with zipfile.ZipFile(path, 'r') as zip_ref:
      zip_ref.extractall(extract_path)

  print(os.listdir(extract_path))

  category = os.listdir(extract_path)
  images = []
  labels = []

  columns = category
  Labelss = category

  count = 0
  for folder in columns:
      count2 = 0
      print('Loading --> ', folder)
      for img in os.listdir(os.path.join(extract_path,folder)):
          image = cv2.imread(os.path.join(extract_path,folder,img))
          image = cv2.resize(image, (128, 128))
          images.append(image)
          labels.append(count)
          count2 += 1
      print('    Total of --> ', count2)
      count += 1

  print(len(images))
  print(len(labels))


  unique_labels = list(set(labels))
  print(unique_labels)

  training_data = []
  for img, label in zip(images, labels):
      training_data.append([img, label])

  X = []
  y = []

  for img , lab in training_data:
      X.append(img)
      y.append(lab)

  X = np.array(X).astype(float)/255
  y = np.array(y)
  X.shape

  model_CNN = tf.keras.models.Sequential([
      tf.keras.layers.Conv2D(32, (3,3), activation ='relu', input_shape =(128,128,3)),
      tf.keras.layers.MaxPooling2D(2,2),
      tf.keras.layers.Conv2D(64, (3,3), activation ='relu'),
      tf.keras.layers.MaxPooling2D(2,2),
      tf.keras.layers.Conv2D(128, (3,3), activation ='relu'),
      tf.keras.layers.MaxPooling2D(2,2),

      tf.keras.layers.Dropout(0.5),
      tf.keras.layers.Flatten(),
      tf.keras.layers.Dense(250, activation='relu'),
      tf.keras.layers.Dense(2, activation='softmax'),
  ])

  model_CNN.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

  TensorBoardCNN = TensorBoard(log_dir='logs/denso')

  model_CNN.fit(X,y, batch_size= 32, validation_split=0.15, epochs=100, callbacks=[TensorBoardCNN])
# Plot the accuracy curve
  plt.figure(figsize=(10, 10))
  plt.xlabel("# Epochs")
  plt.ylabel("Loss")
  plt.plot(model_CNN.history.history['loss'])
  plt.savefig(f"loss_plot_{fruit}.png")

  files.download(f"/content/loss_plot_{fruit}.png")

  model_CNN.save(f"model_CNN_{fruit}.h5")

  files.download(f"/content/model_CNN_{fruit}.h5")
```
